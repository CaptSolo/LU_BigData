# Titanic dataset — Plotly visualizations

Interactive exploration of the classic Titanic passenger dataset using [Plotly Express](https://plotly.com/python/plotly-express/).

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio

# necessary in order to show visualizations in a Jupyter notebook
pio.renderers.default = 'notebook'
```

## Load the dataset

We use the Titanic dataset from the `seaborn-data` repository. Each row is a passenger.

```python
# Load the Titanic dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
titanic = pd.read_csv(url)
```

```python
titanic.head(5)
```

```python
titanic.shape
```

## histogram() — Age distribution

https://plotly.com/python/histograms/

```python
fig = px.histogram(
    titanic,
    x='age',
    color='sex',
    nbins=30,
    barmode='overlay',
    opacity=0.7,
    title='Age distribution of Titanic passengers by gender',
    height=500
)
fig
```

## bar() — Survival rate by class and gender

https://plotly.com/python/bar-charts/

```python
survival = (
    titanic.groupby(['class', 'sex'])['survived']
    .mean()
    .reset_index()
    .rename(columns={'survived': 'survival_rate'})
)
survival
```

```python
fig = px.bar(
    survival,
    x='class',
    y='survival_rate',
    color='sex',
    barmode='group',
    text_auto='.0%',
    labels={'survival_rate': 'Survival rate'},
    title='Survival rate by passenger class and gender',
    height=500
)
fig
```

## box() — Fare distribution by class

https://plotly.com/python/box-plots/

```python
fig = px.box(
    titanic,
    x='class',
    y='fare',
    color='survived',
    points='outliers',
    category_orders={'class': ['First', 'Second', 'Third']},
    labels={'survived': 'Survived', 'fare': 'Fare (£)'},
    title='Fare distribution by class and survival',
    height=500
)
fig
```

## scatter() — Age vs. fare

https://plotly.com/python/line-and-scatter/

```python
fig = px.scatter(
    titanic.dropna(subset=['age', 'fare']),
    x='age',
    y='fare',
    color='class',
    symbol='survived',
    hover_data=['sex', 'embark_town'],
    labels={'age': 'Age (years)', 'fare': 'Fare (£)'},
    title='Age vs. fare, coloured by class',
    height=600
)
fig
```

## sunburst() — Passenger breakdown

Hierarchical view: class → sex → survival.

https://plotly.com/python/sunburst-charts/

```python
sunburst_data = titanic.copy()
sunburst_data['survived_label'] = sunburst_data['survived'].map({0: 'Died', 1: 'Survived'})

fig = px.sunburst(
    sunburst_data,
    path=['class', 'sex', 'survived_label'],
    color='survived_label',
    color_discrete_map={'Survived': '#2ecc71', 'Died': '#e74c3c'},
    title='Passenger breakdown: class → sex → outcome',
    height=600
)
fig
```
