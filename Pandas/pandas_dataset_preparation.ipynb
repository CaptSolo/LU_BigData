# Generating Synthetic Research Publication Data

In this notebook we'll generate a synthetic dataset of ~50,000 journal articles
and save the result to a CSV file.

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

np.random.seed(42)

N = 50000

fields_clean = ["Computer Science", "Biology", "Physics", "Economics", "Psychology", "Chemistry"]

# Some messy data for the data cleaning exercise
fields_messy = fields_clean + ["CS", "Comp Sci", "Bio", "Econ", "Psych", "Chem"]

articles = pd.DataFrame({
    "article_id":    range(1, N + 1),
    "title":         [f"Article_{i}" for i in range(1, N + 1)],
    "field":         np.random.choice(fields_messy, N, p=[
                         0.14, 0.14, 0.14, 0.14, 0.14, 0.14,   # clean names
                         0.03, 0.02, 0.02, 0.02, 0.02, 0.05    # messy variants
                     ]),
    "pub_date":      pd.to_datetime(
                         np.random.choice(pd.date_range("2015-01-01", "2024-06-30"), N)
                     ),
    "citations":     np.random.lognormal(mean=2.5, sigma=1.4, size=N).astype(int),
    "author_count":  np.random.poisson(lam=4, size=N) + 1,
    "open_access":   np.random.choice([True, False], N, p=[0.45, 0.55]),
    "funding_source": np.random.choice(
                         ["Government", "Industry", "University", "Foundation", np.nan],
                         N, p=[0.30, 0.15, 0.20, 0.10, 0.25]
                     ),
})
```

```python
print(f"Dataset shape: {articles.shape}")
articles.head(10)
```

```python
articles.to_csv("pandas_dataset.csv")
```
